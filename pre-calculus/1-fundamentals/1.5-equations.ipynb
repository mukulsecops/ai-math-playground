## Ques 69-74. Use the discriminant to determine the number of real solutions of the equation. Do not solve the equation

```python
def det_real_sols(eqs, ques_start):
    for i, eq in enumerate(eqs, ques_start):
        display(Math(f"{i}: {latex(eq)}"))
        dis = discriminant(eq)
        if dis > 0:
            print(f"\t{dis} has two distinct real solutions")
        elif dis < 0:
            print(f"\t{dis} has exactly one real solutions")
        else:
            print(f"\t{dis} has no real solutions")

eqs_6 = [
    Eq(x**2 - 6*x + 1, 0),
    Eq(3*x**2, 6*x - 9),
    Eq(x**2 + 2.20*x + 1.21, 0),
    Eq(x**2 + 2.21*x + 1.21, 0),
    Eq(4*x**2 + 5*x + Rational(13,8), 0),
    Eq(x**2 + 2*x - 12, 0),
]

det_real_sols(eqs_6, 69)
            
```

### Ques 99-100 . Falling-Body Problems

Suppose an object is dropped from a height h0 above the ground. Then its height after
t seconds is given by h = -16t^2 + h0, where h is measured in
feet. Use this information to solve the problem.

99. If a ball is dropped from 288 ft above the ground, how long does it take to reach ground level?

100. A ball is dropped from the top of a building 96 ft tall.
        * How long will it take to fall half the distance to ground level?
        * How long will it take to fall to ground level?

```python
fall_eq = Eq(h, -16*t**2 + h0)

#99 - Height on ground level is 0 so h = 0 and given h0 = 288 ft
solveset(fall_eq.subs({h0: 288, h:0}).evalf(), t)
# Hence the solution is 4.24 seconds

#100- a) if t=0 seconds then h=h0 so to find half the distance h=h0/2
solveset(fall_eq.subs({h0: 96, h:96/2}).evalf(), t)
# Hence the solution is 1.73 seconds.

#100 - b) repeating q:99 logic hence h=0 on ground level
solveset(fall_eq.subs({h0: 96, h:0}).evalf(), t)
# hence the solution is 2.44 seconds to fall to ground level.
```

### Ques 101-102 .  Falling-Body Problems
Use the formula h = -16t^2 + v0*t.

101. A ball is thrown straight upward at an initial speed of v0 = 40 ft/s.
        - When does the ball reach a height of 24 ft?
        - When does it reach a height of 48 ft?
        - What is the greatest height reached by the ball?
        - When does the ball reach the highest point of its path?
        - When does the ball hit the ground?

102. How fast would a ball have to be thrown upward to reach a maximum height of 100 ft? [Hint: Use the discriminant of the equation 16t^2 - v0*t + h = 0]

```python
fall_formula = Eq(h, -16*t**2 + v0*t)

# 101 a)   v0=40 ft/s and h=24
solve(fall_formula.subs({v0: 40, h:24}))
# two solution is after 1 sec and after 3/2 seconds

# 101 b)   v0=40 ft/s and h=48
solve(fall_formula.subs({v0: 40, h:48}))
# no solution hence it never reaches that much height.

# 101 c)  to find maximum height we use discriminant where D=0 because it will only yield exactly one solution.
max_height = Eq(0, discriminant(fall_formula.subs({v0: 40})))
solveset(max_height, h)
# Max height ball goes is 25 ft.

# 101 d) Using the maximum height 25 ft we find the time when ball reaches its heightest point.
solve(fall_formula.subs({v0: 40, h:25}), t)
# It reaches 5/4 seconds to heightest point.

# 101 e) hence to reach ground the height must be 0 so h=0 and solve in respect with t.
solve(fall_formula.subs({v0: 40, h:0}), t)


# 102  to find speed v0 we use discriminant where D=0 because it will only yield exactly one solution.
v0_speed = Eq(0, discriminant(fall_formula.subs({h: 100})))
solve(v0_speed, v0)
# speed to reach height h=100 ft is 80 ft/s
```

### Ques 103. Shrinkage in Concrete Beams

As concrete dries, it shrinksâ€”the higher the water content, the greater the
shrinkage. If a concrete beam has a water content of w kg/m^3 then it will shrink by a factor:

S = (0.032w - 2.5)/10000

where S is the fraction of the original beam length that
disappears due to shrinkage.

1. A beam 12.025 m long is cast in concrete that contains
    250 kg/m3 water. What is the shrinkage factor S? How
    long will the beam be when it has dried?

2. A beam is 10.014 m long when wet. We want it to
    shrink to 10.009 m, so the shrinkage factor should be
    S = 0.00050. What water content will provide this
    amount of shrinkage?

```python
srinkage_eq = Eq(S, (0.032*w - 2.5)/10000)

# A:103 - 1. w=250 kg/m3
shrinkage_fr = solve(srinkage_eq.subs({w: 250}), S)[0]
# S = 0.00055
# How long will the beam be when it has dried?
12.025 - (12.025 * shrinkage_fr)

# A:103 - 2. S = 0.00050
solve(srinkage_eq.subs({S: 0.00050}), w)
# Amount of water content should be provided 234.375 kg/m3
```

### Ques-105 . Fish Population

The fish population in a certain lake
rises and falls according to the formula

F = 1000*(30 + 17*t + t**2)

Here F is the number of fish at time t, where t is measured
in years since January 1, 2002, when the fish population
was first estimated.

1. On what date will the fish population again be the
    same as on January 1, 2002?

```python
fish_pop_eq = Eq(F, 1000*(30 + 17*t + t**2))

#105 - 1) First find F on Jan 1,2002 means t=0 then put F and find val of t again
solve(fish_pop_eq.subs({t: 0}), F) # we got F here.

solve(fish_pop_eq.subs({F:30000}), t)
# Hence the date is after 17 years i.e Jan 1, 2019


# 105 - 2) To find date on which all fishes will die (F=0)
solve(fish_pop_eq.subs({F:0}), t)
```

### Ques 109. Depth of a Well

One method for determining the depth
of a well is to drop a stone into it and then measure the
time it takes until the splash is heard.

If d is the depth of
the well (in feet) and t1 the time (in seconds) it takes
for the stone to fall, then

d = 16*t1**2

so t1 = sqrt(d)/4. 

Now if t2 is the time it takes for the sound to travel back up, then 

d = 1090*t2 because the speed of sound is 1090 ft/s.

So t2 = d/1090. Thus, the total time elapsed between dropping the stone and hearing the splash is

t1 + t2 = sqrt(d)/4 + d/1090

How deep is the well if this total time is 3 s?



```python
total_time_elapsed = Eq(T, (sqrt(d)/4) + (d/1090))
solve(total_time_elapsed.subs({T:3}).evalf(), d)

# Hence the answer is 132.6 ft
```

### Ques 110. A Family of Equations

The equation

3*x + k - 5 = k*x - k + 1

is really a family of equations, because for each value
of k, we get a different equation with the unknown x.

The letter k is called a **parameter** for this family. What value
should we pick for k to make the given value of x a solution of the resulting equation?

1. x = 0
2. x = 1
3. x = 2

```python
family_eq = Eq(3*x + K - 5, K*x - K + 1)
solve(family_eq, x) # Hence we should pick x=2 for finding the value of k

solve(family_eq.subs({x: 0}), K)
# Hence value of K should be 3
```

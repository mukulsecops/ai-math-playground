# Equations excercises 1.5

```python
from sympy import *
from IPython.display import Math
```

```python
x,y,z,w,t = symbols('x,y,z,w,t')

def det_solution(eq, vals):
    display(Math(f"For eq: {latex(eq)}"))
    for val in vals:
        print(f"\tval {val} is {eq.subs({x:val})}")


def gen_equations(eqs, ques_start):
    for i, eq in enumerate(eqs, start=ques_start):
        display(Math(f"{i}: {latex(eq)}"))
```

## Ques. 1-4. Determine whether the given value is a solution of the equation.

```python
eq1 = Eq(4*x + 7, 9*x - 3)
det_solution(eq1, [-2,2]) # Is the solution LHS=RHS

eq2 = Eq(1-(2-(3-x)), 4*x - (6+x))
det_solution(eq2, [2,4]) # 2 is the solution

eq3 = Eq((1/x) - (1/(x-4)), 1)
det_solution(eq3, [2,4])

eq4 = Eq((x**Rational(3,2))/(x - 6), x - 8)
det_solution(eq4, [4,8])
```

## Ques. 5-22. The given equation is either linear or equivalent to a
### linear equation. Solve the equation.

```python
eqs_1 = [
    solve(Eq(2*x + 7, 31)),
    solve(Eq(5*x - 3, 4)),
    solve(Eq(Rational(1,2)*x - 8, 1)),
    solve(Eq(3+Rational(1,3)*x, 5)),
    solve(Eq(-7*w, 15 - 2*w)),
    solve(Eq(5*t - 13, 12 - 5*t)),
    solve(Eq(Rational(1,2)*y - 2, Rational(1,3)*y)),
    solve(Eq(z/5, Rational(3,10)*z + 7)),
    solve(Eq(2*(1 - x), 3*(1 + 2*x) + 5)),
    solve(Eq(Rational(2,3)*y + Rational(1,2)*(y - 3), (y + 1)/4)),
    solve(Eq(x - Rational(1,3)*x - Rational(1,2)*x - 5, 0)),
    solve(Eq(2*x - x/2 + (x+1)/4, 6*x)),
    solve(Eq(1/x,(4/(3*x)) + 1)),
    solve(Eq((2*x - 1)/(x+2), Rational(4,5))),
    solve(Eq((3/(x+1)) - Rational(1,2), 1/(3*x + 3))),
    solve(Eq((4/(x-1)) + (2/(x+1)), 35/(x**2 - 1))),
    solve(Eq((t - 4)**2, (t + 4)**2 + 32)),
    solve(Eq(sqrt(3)*x + sqrt(12), (x + 5)/sqrt(3))),
]

gen_equations(eqs_1, 5)
```

## Ques 23-36. Solve the equation for the indicated variable.

```python
P,V,n,R,T,F,G,M,m,r,R1,R2,l,a,b,c,d,V,h,A,i,g,v0,S = symbols("P,V,n,R,T,F,G,M,m,r,R1,R2,l,a,b,c,d,V,h,A,i,g,v0,S")

eqs_2 = [
    solve(Eq(P*V, n*R*T), R),
    solve(Eq(F,G*((m*M)/r**2)), m),
    solve(Eq(1/R, 1/R1 + 1/R2), R1),
    solve(Eq(P, 2*l + 2*w), w),
    solve(Eq((a*x + b)/(c*x + d), 2), x),
    solve(Eq(a - 2*(b - 3*(c - x)), 6), x),
    solve(Eq(a**2*x + (a - 1), (a + 1)*x), x),
    solve(Eq((a+1)/b, ((a-1)/b) + ((b+1)/a)), a),
    solve(Eq(V, Rational(1,3)*pi*r**2*h), r),
    solve(Eq(F,G*((m*M)/r**2)), r),
    solve(Eq(a**2 + b**2, c**2), b),
    solve(Eq(A, P*((1 + i/100)**2)), i),
    solve(Eq(h,Rational(1,2)*g*t**2 + v0*t),t),
    solve(Eq(S, Rational(1,2)*(n*(n + 1))), n)
]

gen_equations(eqs_2, 23)
```

## Ques 37-44. Solve the equation by factoring.

## Ques 44-52. Solve the equation by completing the square.

```python
eqs_4 = [
    solve(Eq(x**2 + 2*x - 5, 0)),
    solve(Eq(x**2 - 4*x + 2, 0)),
    solve(Eq(x**2 + 3*x - Rational(7,4), 0)),
    solve(Eq(x**2, Rational(3,4)*x - Rational(1,8))),
    solve(Eq(2*x**2 + 8*x + 1, 0)),
    solve(Eq(3*x**2 - 6*x - 1, 0)),
    solve(Eq(4*x**2 - x, 0)),
    solve(Eq(-2*x**2 + 6*x + 3, 0)),
    
]

gen_equations(eqs_4, 45)
det_solution(Eq(-2*x**2 + 6*x + 3, 0), solve(Eq(-2*x**2 + 6*x + 3, 0))) # Checking solutions are true for all equations.
```

```python
from sympy import *
from IPython.display import Math
```

```python
x,y,z,w,t = symbols('x,y,z,w,t', real=True)

def det_solution(eq, vals):
    display(Math(f"For eq: {latex(eq)}"))
    for val in vals:
        print(f"\tval {val} is {eq.subs({x:val})}")


def gen_equations(eqs, ques_start):
    for i, eq in enumerate(eqs, start=ques_start):
        display(Math(f"{i}: {latex(eq)}"))
```

## Ques. 1-4. Determine whether the given value is a solution of the equation.

```python
eq1 = Eq(4*x + 7, 9*x - 3)
det_solution(eq1, [-2,2]) # Is the solution LHS=RHS

eq2 = Eq(1-(2-(3-x)), 4*x - (6+x))
det_solution(eq2, [2,4]) # 2 is the solution

eq3 = Eq((1/x) - (1/(x-4)), 1)
det_solution(eq3, [2,4])

eq4 = Eq((x**Rational(3,2))/(x - 6), x - 8)
det_solution(eq4, [4,8])
```

## Ques 69-74. Use the discriminant to determine the number of real solutions of the equation. Do not solve the equation

```python
def det_real_sols(eqs, ques_start):
    for i, eq in enumerate(eqs, ques_start):
        display(Math(f"{i}: {latex(eq)}"))
        dis = discriminant(eq)
        if dis > 0:
            print(f"\t{dis} has two distinct real solutions")
        elif dis < 0:
            print(f"\t{dis} has exactly one real solutions")
        else:
            print(f"\t{dis} has no real solutions")

eqs_6 = [
    Eq(x**2 - 6*x + 1, 0),
    Eq(3*x**2, 6*x - 9),
    Eq(x**2 + 2.20*x + 1.21, 0),
    Eq(x**2 + 2.21*x + 1.21, 0),
    Eq(4*x**2 + 5*x + Rational(13,8), 0),
    Eq(x**2 + 2*x - 12, 0),
]

det_real_sols(eqs_6, 69)
            
```

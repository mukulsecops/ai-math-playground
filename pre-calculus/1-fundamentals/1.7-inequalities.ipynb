98. **Gas Mileage** The gas mileage g (measured in mi/gal) for

g = 10 + 0.9v - 0.01v**2

, as long as v is between 10 mi/h
and 75 mi/h. For what range of speeds is the vehicles
mileage 30 mi/gal or better?

```python
# Set up the equation
g,v = symbols('g,v')

gas_mileage_eq = Eq(g, 10 + 0.9*v - 0.01*v**2)

# 10 < v < 75

solveset(solve(gas_mileage_eq, g)[0] >= 30, v, domain=Interval(-oo, oo))
```

102. **Manufacturers Profit** If a manufacturer sells x units of
a certain product, his revenue R and cost C (in dollars) are
given by:

- R = 20x
- C = 2000 + 8x + 0.0025x**2

Use the fact that:

profit = revenue - cost

to determine how many units he should sell to enjoy a
profit of at least $2400.

```python
# Set up the equation
R = symbols('R')

revenue_x_units = Eq(R, 20*x)
cost_x_units = Eq(C, 2000 + 8*x + 0.0025*x**2)

profit_x = solve(revenue_x_units, R)[0] - solve(cost_x_units, C)[0]

reduce_inequalities(profit_x >= 2400, x)
```

104.  **Airline Ticket Price** A charter airline finds that on its
Saturday flights from Philadelphia to London, all 120
seats will be sold if the ticket price is $200. However, for
each $3 increase in ticket price, the number of seats sold
decreases by one.

- Find a formula for the number of seats sold if the
ticket price is P dollars

- Over a certain period, the number of seats sold for this
flight ranged between 90 and 115. What was the corresponding range of ticket prices?

```python
# Change in price.

P = symbols('P')

price_change  = P - 200
# For every $3 increase the number of seats decrease by 1
total_decrease = price_change/ 3

# Subtract the decrease from number of seats
number_of_seats = 120 - total_decrease

reduce_inequalities([90 < number_of_seats, number_of_seats < 115], P)
```

106. **Fencing a Garden** A determined gardener has 120 ft

vegetable garden in her backyard, and she wants the area

possible for the length of her garden?

```python
# Let l be the length of her garden
l,w = symbols('l,w')

# Express all in terms of l
perimeter = Eq(P, 2*(l + w))
width_garden = solve(perimeter.subs({P:120}), w)[0]

# Set up the model
reduce_inequalities((l * width_garden) >= 800, l)
```

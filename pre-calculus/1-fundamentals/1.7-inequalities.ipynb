## Ques 1-6

 Let S = {-2,-1,0,1/2, 1, sqrt(2), 2, 4} . Determine which
elements of S satisfy the inequality

```python
# Define S
NumSet = {-2, -1, 0, Rational(1,2), 1, sqrt(2), 2, 4}

exprs_1 = [
    3 - 2*x <= Rational(1,2),
    2*x - 1 >= x,
    And(1 < 2*x - 4, 2*x - 4 <= 7), # Equavalent to 1 < 2x - 4 <= 7
    And(-2 <= 3 - x, 3 - x < 2),
    # 1/x <= 1/2 This will not work for value 0 as it will come undefined.
    x**2 + 2 < 4
]

for i, expr in enumerate(exprs_1, start=1):
    for item in NumSet:
        display(Math(f"Q{i}: {item}: {latex(expr.subs({x: item}))}"))
```

## Ques 77-80

A phrase describing a set of real numbers is given.
Express the phrase as an inequality involving an absolute value.

```python
# 77. All real numbers x less than 3 units from 0
reduce_abs_inequality(Abs(x) - 3, '<', x).as_set()

# 78. All real numbers x more than 2 units from 0
reduce_abs_inequality(Abs(x) - 2, '>', x).as_set()

# 79. All real numbers x at least 5 units from 7
reduce_abs_inequality(Abs(x - 7) - 5, '>=', x).as_set()

# 80. All real numbers x at most 4 units from 2
reduce_abs_inequality(Abs(x - 2) - 4, '<=', x).as_set()
```
